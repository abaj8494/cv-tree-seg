## statistical structure

```python
# plotting statistical structure of all the images
# checking max / min / mean / std of the images
# using nrg and rgb images:
def plot_statistics(image, title):
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 4, 1)
    plt.imshow(image)
    plt.title(title)
    
    plt.subplot(1, 4, 2)
    plt.hist(image.ravel(), bins=256, color='gray', alpha=0.7)
    plt.title('Histogram')
    
    plt.subplot(1, 4, 3)
    plt.imshow(np.mean(image, axis=2), cmap='gray')
    plt.title('Mean Image')
    
    plt.subplot(1, 4, 4)
    plt.imshow(np.std(image, axis=2), cmap='gray')
    plt.title('Std Dev Image')
    
    plt.tight_layout()
    plt.show()
# Plot statistics for RGB image
rgb_image = cv2.imread(prefix + "RGB_images/RGB_ar037_2019_n_06_04_0.png")
plot_statistics(rgb_image, "RGB Image")
# Plot statistics for NRG image
nrg_image = cv2.imread(prefix + "NRG_images/NRG_ar037_2019_n_06_04_0.png")
plot_statistics(nrg_image, "NRG Image")
```
